```python
sns.pairplot(data=df,height=3,aspect=1.5)
```

```python
plt.plot(df['National Income'],'or',label='National Income')
plt.plot(df['Per person'],'^k',label='Per person')
plt.legend()
plt.title('National Income Vs Per person')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
```

```python
sns.jointplot(y='National Income',x='Per person',data=df,kind='reg',height=10,color='k')
```

```python
cmap=sns.cubehelix_palette(as_cmap=True,light=1.5,dark=0,reverse=True)

sns.jointplot(y='National Income',x='Per person',data=df,kind='kde',height=10,color='k',cmap=cmap)
sns.jointplot(y='National Income',x='Per person',data=df,kind='hex',height=10,color='k',cmap=cmap)
```

```python
cmap=sns.cubehelix_palette(as_cmap=True,light=1,dark=0,reverse=True,n_colors=10)
g=sns.PairGrid(data=df,height=3,aspect=1.4)
g.map_diag(sns.kdeplot,shade=True,color='r')
g.map_offdiag(sns.kdeplot,shade=True,cmap=cmap)
```
